## 5. üßπ Data Cleaning & Preprocessing

```python
# Remove duplicates
print("Duplicates before:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicates after:", df.duplicated().sum())
```

### Handlig Missing Values

```python
# 1. Check missing values BEFORE
print("Missing values per column BEFORE filling:\n")
print(df.isna().sum())
```

```python
# 2. Handle missing values
numeric_cols = df.select_dtypes(include=np.number).columns
categorical_cols = df.select_dtypes(exclude=np.number).columns

# Numeric columns ‚Üí fill with median
for col in numeric_cols:
    median_val = df[col].median()
    df[col] = df[col].fillna(median_val)

# Categorical columns ‚Üí fill with mode
for col in categorical_cols:
    if not df[col].mode().empty:   # check if mode exists
        mode_val = df[col].mode()[0]
        df[col] = df[col].fillna(mode_val)
```

```python
# 3. Check missing values AFTER
print("\nMissing values per column AFTER filling:\n")
print(df.isna().sum())
```

```python
# 4. Summary table (Before vs After)
missing_summary = pd.DataFrame({
    "Missing Before": df.isna().sum(),
    "Missing After": df.isna().sum()
})
print("\nüìä Missing Value Summary (Before vs After):\n")
print(missing_summary)
```

```python
# Missing Values are none
```

```python
#Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['quarter'] = df['date'].dt.quarter

# Display with emojis in column names
df = df.rename(columns={
    'date': 'üìÖ Date',
    'year': 'üìÜ Year',
    'month': 'üåô Month',
    'quarter': 'üü¶ Quarter'
})

df[['üìÖ Date', 'üìÜ Year', 'üåô Month', 'üü¶ Quarter']]
```

## 6. ‚öôÔ∏è Feature Engineering

```python
# Net units sold
df["üì¶ Net Units Sold"] = df["units_sold"] - df["return_units"]

# Effective price after discount
df["üí∞ Effective Unit Price"] = df["revenue"] / df["units_sold"]

# Profit margin (only if profit column exists)
if "profit" in df.columns:
    df["üìä Profit Margin (%)"] = (df["profit"] / df["revenue"]) * 100
```

```python
# List of engineered columns
engineered_cols = ["üì¶ Net Units Sold", "üí∞ Effective Unit Price"]

# Only add profit column if it exists
if "üìä Profit Margin (%)" in df.columns:
    engineered_cols.append("üìä Profit Margin (%)")

# Display Result
Result = df[engineered_cols].sample(50, random_state=42).style.format({
    "üí∞ Effective Unit Price": "{:.2f}",
    **({"üìä Profit Margin (%)": "{:.2f}%"} if "üìä Profit Margin (%)" in df.columns else {})
})

# Only apply gradient if profit column exists
if "üìä Profit Margin (%)" in df.columns:
  Result = Result.background_gradient(subset=["üìä Profit Margin (%)"], cmap="RdYlGn")

Result
```

## 7. üîç Exploratory Data Analysis (EDA)

### Univariate:

```python
df["units_sold"].hist(bins=30)
plt.title("Distribution of Units Sold")
plt.xlabel("Units Sold")
plt.ylabel("Frequency")
plt.show()
```

üìä **Univariate Analysis ‚Äì Units Sold**

- The distribution of units sold appears fairly **uniform** between 50 and 300.  
- A **noticeable cluster** exists for products with very low sales (0‚Äì50 units).  
- Unlike a normal distribution, sales are spread across multiple ranges without a single dominant peak.  
- This suggests that demand is **diverse across products**, with some selling in small batches and others consistently at higher volumes.


### Bivariate:

```python
df.groupby("product_type")["revenue"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
plt.title("Revenue by Product Type")
plt.ylabel("Revenue")
plt.show()
```

üìä **Bivariate Analysis ‚Äì Revenue by Product Type**

- The analysis shows that **DC Valve** is the **top revenue generator**, contributing the majority of total sales.  
- Other product types such as **External Gear Pump** and **DCV Valve** contribute relatively less in comparison.  
- This indicates that **revenue is highly concentrated** in a few product categories, with DC Valve being the core driver.  
- Business implication: Maintaining supply and improving margins for DC Valves should be a **strategic priority**, while opportunities for growth may lie in boosting underperforming product lines.

### Multivariate:

```python
corr = df.corr(numeric_only=True)
plt.imshow(corr, cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```

üìä **Multivariate Analysis ‚Äì Correlation Heatmap**

- **Revenue** is strongly correlated with **Units Sold** and **Net Units Sold**, confirming that sales volume is the primary driver of revenue.  
- **Discount Percent** shows a negative relationship with **Effective Unit Price**, which is expected since higher discounts reduce per-unit revenue.  
- Other external factors such as **competitor activity, seasonality index, failure rate** show weaker correlations, suggesting they have a limited direct impact on revenue in this dataset.  
- **Business Implication:** Focusing on increasing units sold while optimizing discounts can maximize revenue, whereas competitor and seasonal effects appear less influential here.

## 8. üìä Visualizations

```python
import seaborn as sns
```

```python
# 1. Monthly Revenue Trend
monthly_revenue = df.groupby('üåô Month')['revenue'].sum()
plt.figure(figsize=(10,5))
monthly_revenue.plot(marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
print("üìà Insight: Revenue shows fluctuations across months with some seasonal peaks.\n")
```

```python
# Create Profit and Profit Margin
df['Profit'] = df['revenue'] - df['marketing_spend']
df['Profit_Margin'] = (df['Profit'] / df['revenue']) * 100

# 2 .Monthly Profit Trend plot
monthly_profit = df.groupby('üåô Month')['Profit'].sum()
plt.figure(figsize=(10,5))
monthly_profit.plot(marker='o', color='green')
plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
print("üí∞ Insight: Profit follows a similar trend to revenue but with sharper variations.\n")
```

```python
# 3. Revenue by Channel
plt.figure(figsize=(8,5))
df.groupby('channel')['revenue'].sum().plot(kind='bar', color='orange')
plt.title("Revenue by Channel")
plt.xlabel("Channel")
plt.ylabel("Total Revenue")
plt.show()
print("üì¶ Insight:Revenue generation is fairly balanced across all channels, with no single channel showing clear dominance.\n")
```

```python
# 4. Revenue by Region
plt.figure(figsize=(8,5))
df.groupby('region')['revenue'].sum().sort_values().plot(kind='bar')
plt.title("Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.show()
print("üåç Insight:Revenue is evenly distributed across all regions, with no region showing a major dominance.\n")
```

```python
# 5. Revenue by Product Type
plt.figure(figsize=(8,5))
df.groupby('product_type')['revenue'].sum().sort_values().plot(kind='bar')
plt.title("Revenue by Product Type")
plt.xlabel("Product Type")
plt.ylabel("Total Revenue")
plt.show()
print("üõ†Ô∏è Insight: Certain product types like valves dominate revenue compared to others.\n")
```

```python
# 6.Units Sold
plt.figure(figsize=(8,5))
df['units_sold'].hist(bins=30, color='skyblue')
plt.title("Distribution of Units Sold")
plt.xlabel("Units Sold")
plt.ylabel("Frequency")
plt.show()
print("üìä Insight: Most products fall in a mid-range of units sold, with fewer extreme outliers.\n")
```

```python
# 7. Revenue
plt.figure(figsize=(8,5))
df['revenue'].hist(bins=30, color='purple')
plt.title("Distribution of Revenue")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
print("üíπ Insight: Revenue distribution is skewed, with a majority of sales clustered in lower ranges.\n")
```

```python
# 8. Profit Margin by Segment
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='customer_segment', y='Profit_Margin', palette='Set2')
plt.title("Profit Margin by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Profit Margin (%)")
plt.show()
print("üì¶ Insight: Profit margins are generally high across all customer segments (~90%), but Distributor and Retail segments show more variability and negative outliers compared to OEM and Aftermarket..\n")
```

```python
# 9. Revenue vs Marketing Spend
plt.figure(figsize=(8,5))
plt.scatter(df['marketing_spend'], df['revenue'], alpha=0.5, color='red')
plt.title("Revenue vs Marketing Spend")
plt.xlabel("Marketing Spend")
plt.ylabel("Revenue")
plt.show()
print("üì¢ Insight: Marketing spend does not show a strong linear relationship with revenue. While higher spend sometimes aligns with higher revenue, there are many cases where high revenue occurs even with low spend, indicating other factors also drive revenue..\n")
```
